```python
data = pd.read_excel("population.xlsx")
data["年份"] = pd.to_datetime(data["年份"],format="%Y")
data = data.set_index("年份",drop=True)
data = data[~data.index.duplicated(keep="first")]

data = data.loc["1990-1-1":"2019-1-1",:]
```
